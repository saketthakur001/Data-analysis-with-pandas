```python
import pandas as pd
import math
```

```python
def nCr(n, k):
    f = math.factorial
    return f(n) / (f(k)*f(n-k))
nCr(5, 3)
```

```python
def birthday_prob(num):

    if num < 2: return "Do you think you are funny!!!"
    return 1 - (364 / 365) ** nCr(num, 2)
```

```python
birthday_prob(4)
```
